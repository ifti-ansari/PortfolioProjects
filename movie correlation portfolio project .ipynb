```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'C:\Users\ansar\Downloads\movies.csv')
    
```

```python
# lets have a look at data

df.head()
```

```python
# lets see if there is any missing data 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# data types for columns

df.dtypes
```

```python
#changing column data type for example budget float to int, gross float to int.

df['budget']=df['budget'].astype('Int64')
df['gross']=df['gross'].astype('Int64')
```

```python
df.dtypes

df.head()
```

```python
# sorting gross value in dscending order
df=df.sort_values(by=['gross'], inplace = False , ascending = False)
```

```python
# correlation 
# Budget vs gross
#plt.scatter(x=df['budget'],y=df['gross'])
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs gross Earning')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
#seaborn
#plot budget vs gross

sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"Blue"})
```

```python
#looking at correlation (its only works on numeric data)
# type of correlation #pearson, # kendall , #spearman


df.corr(method='pearson')

```

```python
correlation_matrix=df.corr(method='pearson')

sns.heatmap(correlation_matrix,annot=True)

plt.title('Correlation matrix For numeric Feature')
plt.xlabel('Movie feature')
plt.ylabel('Movie Feature')

plt.show()
```

```python
df_numerized = df


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
df_numerized.corr()
```

```python
#using unstacking 

correlation_mat = df_numerized.corr()
corr_pairs=correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs)>0.4]

high_corr
```
